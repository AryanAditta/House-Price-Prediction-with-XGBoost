# **Importing the Dependencies :**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

# **Importing the Boston House Price Dataset :**

```python
house_price = pd.read_csv('/content/Boston-house-price-data.csv')
```

```python
house_price.head()
```

```python
house_price_df = house_price.drop('MEDV', axis=1)
```

```python
house_price_df.head()
```

```python
house_price_df['Price']=house_price.MEDV
```

```python
house_price_df.head()
```

```python
house_price_df.shape
```

```python
house_price_df.isnull().sum()
```

```python
house_price_df.describe()
```

Understanding the correlation between various features in the dataset

```python
correlation = house_price_df.corr()
```

```python
# constructing a heatmap to nderstand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

# **Splitting the data and the Target :**

```python
house_price_df.head()
```

```python
X = house_price_df.drop(['Price'], axis=1)
Y = house_price_df['Price']
```

```python
print(X)
print(Y)
```

# **Splitting the data into Training data and Test data :**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Model Training :**

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

# **Prediction on training data :**

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(Y_train, training_data_prediction, color='blue', label='Actual vs Predicted (Training)')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price (Training)")
plt.legend()
plt.show()
```

# **Prediction on Test Data :**

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
# Visualizing actual vs predicted prices for test data
plt.figure(figsize=(8, 6))
plt.scatter(Y_test, test_data_prediction, color='red', label='Actual vs Predicted (Test)')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price (Test)")
plt.legend()
plt.show()
```
